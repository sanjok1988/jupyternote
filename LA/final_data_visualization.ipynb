```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file into a DataFrame
df = pd.read_csv('output/final_student_cleaned_data.csv')

# Step 2: Group the DataFrame by the 'class_id' column
grouped = df.groupby('class_id')

# Step 3: Count the number of users in each class
class_counts = grouped.size()

# Step 4: Plot a graph
class_counts.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Number of Users in Each Class')
plt.xlabel('Class ID')
plt.ylabel('Number of Users')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
book_to_class_id_mapping = {
    1750:'English Advantage - Starter Book 1',
    1751:'English Advantage - Novice Book 2',
    1752:'English Advantage - Novice Book 3',
    1753:'English Advantage - Intermediate Book 4',
    1754:'English Advantage - Intermediate Book 5',
    1755:'English Advantage - Intermediate Book 6',
    1756:'English Advantage - Pre-Advanced Book 7',
    1757:'English Advantage - Pre-Advanced Book 8',
    1758:'English Advantage - Pre-Advanced Book 9',
    1759:'English Advantage - Pre-Advanced Book 10',
    1760:'English Advantage - Foundations',
    1761:'Business Edge - Book 1 (A1)',
    1762:'Business Edge - Book 2 (A2)',
    1763:'Business Edge - Book 3 (A2)',
    1764:'Business Edge - Book 4 (B1)',
    1765:'Business Edge - Book 5 (B1)',
    1766:'Business Edge - Book 6 (B1)',
    1767:'Business Edge - Foundations'
}
```

```python
# Mapping class IDs to class names
df['class_name'] = df['class_id'].map(book_to_class_id_mapping)

# Grouping the DataFrame by class_name and counting the number of users in each class
class_counts = df['class_name'].value_counts().sort_index()

# Plotting the graph
plt.figure(figsize=(10, 6))
class_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Users in Each Class')
plt.xlabel('Class Name')
plt.ylabel('Number of Users')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
