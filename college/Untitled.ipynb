```python
import pyautogui
import time

# Define the interval between clicks (in seconds)
click_interval = 1.0

# Function to simulate mouse clicks
def auto_click():
    while True:
        # Perform a left mouse click
        pyautogui.click()
        print('click')
        pyautogui.typewrite('Hello world!')                 # prints out "Hello world!" instantly
        pyautogui.typewrite('Hello world!', click_interval)
        # Wait for the specified interval
        time.sleep(click_interval)

# Call the function to start the auto clicker
auto_click()
```
