#### suppose x has range [0,0.5], y has range [0,1] and f(x,y)=96x^2y^3 then find the marginal probability of x and y. also check their independency

**Given:**

**Joint PDF:** 

\[ f(x,y) = 96x^2y^3 \]

**Range of** \( X \): \( [0,0.5] \)

**Range of** \( Y \): \( [0,1] \)

**Marginal PDF of** \( X \):

To find the marginal PDF of \( X \), we need to integrate the joint PDF \( f(x,y) \) with respect to \( y \) over the range of \( Y \), which is \( [0,1] \).

\[ f_X(x) = \int_{0}^{1} 96x^2y^3 \, dy \]

This integral represents the sum of probabilities along the \( y \)-axis for all possible values of \( y \) given a fixed value of \( x \) within the range of \( X \).

Solving the integral:

\[ f_X(x) = 32x^2 \int_{0}^{1} y^3 \, dy = 32x^2 \left[\frac{1}{4}y^4\right]_{0}^{1} = 8x^2 \]

Thus, the marginal PDF of \( X \) is \( f_X(x) = 8x^2 \).

**Marginal PDF of** \( Y \):

To find the marginal PDF of \( Y \), we integrate the joint PDF \( f(x,y) \) with respect to \( x \) over the range of \( X \), which is \( [0,0.5] \).

\[ f_Y(y) = \int_{0}^{0.5} 96x^2y^3 \, dx \]

Solving the integral:

\[ f_Y(y) = 96y^3 \int_{0}^{0.5} x^2 \, dx = 96y^3 \left[\frac{1}{3}x^3\right]_{0}^{0.5} = 4y^3 \]

Thus, the marginal PDF of \( Y \) is \( f_Y(y) = 4y^3 \).

**Independence Check:**

To check for independence, we multiply the marginal PDFs of \( X \) and \( Y \) together and compare the result with the joint PDF.

\[ f_X(x) \times f_Y(y) = (8x^2) \times (4y^3) = 32x^2y^3 \]

This product should equal the joint PDF \( f(x,y) = 96x^2y^3 \) for all values of \( x \) and \( y \) within their respective ranges.

Since the product matches the joint PDF, the variables \( X \) and \( Y \) are independent.

So, the marginal PDF of \( X \) is \( f_X(x) = 8x^2 \) and the marginal PDF of \( Y \) is \( f_Y(y) = 4y^3 \), and \( X \) and \( Y \) are independent.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define the range of X and Y
range_X = np.linspace(0, 0.5, 100)  # 100 data points between 0 and 0.5
range_Y = np.linspace(0, 1, 100)    # 100 data points between 0 and 1

# Calculate the marginal PDFs f_X(x) and f_Y(y)
f_X = 8 * range_X ** 2
f_Y = 4 * range_Y ** 3

# Plot the marginal PDF of X
plt.figure(figsize=(8, 6))
plt.plot(range_X, f_X, label='Marginal PDF of X', color='blue')
plt.xlabel('X')
plt.ylabel('PDF')
plt.title('Marginal Probability Density Function of X')
plt.grid(True)
plt.legend()
plt.show()

# Plot the marginal PDF of Y
plt.figure(figsize=(8, 6))
plt.plot(range_Y, f_Y, label='Marginal PDF of Y', color='red')
plt.xlabel('Y')
plt.ylabel('PDF')
plt.title('Marginal Probability Density Function of Y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define the joint PDF function
def joint_pdf(x, y):
    return 96 * x**2 * y**3

# Generate a grid of x and y values
x_values = np.linspace(0, 0.5, 100)  # Range of x: [0, 0.5]
y_values = np.linspace(0, 1, 100)     # Range of y: [0, 1]
X, Y = np.meshgrid(x_values, y_values)

# Calculate the corresponding z values using the joint PDF function
Z = joint_pdf(X, Y)

# Plot the contour graph
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, cmap='viridis')
plt.colorbar(contour)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour Graph of Joint PDF')
plt.grid(True)
plt.show()
```
